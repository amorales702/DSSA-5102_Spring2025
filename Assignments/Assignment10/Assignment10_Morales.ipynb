```python

# Load the CSV file
df = pd.read_csv('Terrapin_2024_NJ.csv')

# Check the first few rows of the annotations column to see the format
print(df['annotations'].head(10))

# Convert stringified Python list/dictionary to actual Python objects
def parse_annotation_str(x):
    try:
        # Use ast.literal_eval to convert the string into a Python literal (list of dicts)
        return ast.literal_eval(x) if pd.notnull(x) else []
    except Exception as e:
        print(f"Error parsing annotation: {e}")
        return []  # If parsing fails, return an empty list

# Apply parsing function to 'annotations' column
df['annotations_parsed'] = df['annotations'].apply(parse_annotation_str)

# Print out the parsed annotations to inspect the result
print(df['annotations_parsed'].head(10))  # Check if parsing works

# Extract relevant annotations (Life Stage, Sex, Alive or Dead)
def extract_individual_annotations(annotations):
    # Maps for controlled attribute and value IDs
    attribute_map = {
        1: 'Life Stage',
        9: 'Sex',
        17: 'Alive or Dead'
    }

    value_map = {
        2: 'Adult', 7: 'Egg', 8: 'Juvenile',
        10: 'Female', 11: 'Male',
        18: 'Alive', 19: 'Dead', 20: 'Cannot Be Determined'
    }

    # Initialize a dictionary to store extracted attributes
    extracted_data = {
        'life_stage': None,
        'sex': None,
        'alive_or_dead': None
    }

    # Check if annotations is a valid list
    if isinstance(annotations, list):
        # Loop through annotations to fill the extracted data
        for a in annotations:
            attr_id = a.get('controlled_attribute_id')
            val_id = a.get('controlled_value_id')

            # Assign values to corresponding columns based on the attribute ID
            if attr_id == 1 and val_id in value_map:
                extracted_data['life_stage'] = value_map[val_id]
            elif attr_id == 9 and val_id in value_map:
                extracted_data['sex'] = value_map[val_id]
            elif attr_id == 17 and val_id in value_map:
                extracted_data['alive_or_dead'] = value_map[val_id]

    return extracted_data

# Apply the extraction function to the 'annotations_parsed' column
annotation_columns = df['annotations_parsed'].apply(extract_individual_annotations)

# Expand the extracted data into separate columns
annotation_columns_df = annotation_columns.apply(pd.Series)

# Join the new columns with the original DataFrame
df = df.join(annotation_columns_df)

# Check if the new columns are populated
print(df[['id', 'life_stage', 'sex', 'alive_or_dead']].head())

# Optional: save to new CSV
df.to_csv("Terrapin_2024_NJ_with_lifestage.csv", index=False)
print("Saved to Terrapin_2024_NJ_with_lifestage.csv")
```

```python

# Load the CSV file
df = pd.read_csv('INat_Observations_NJ.csv')

# Check the first few rows of the annotations column to see the format
print(df['annotations'].head(10))

# Convert stringified Python list/dictionary to actual Python objects
def parse_annotation_str(x):
    try:
        # Use ast.literal_eval to convert the string into a Python literal (list of dicts)
        return ast.literal_eval(x) if pd.notnull(x) else []
    except Exception as e:
        print(f"Error parsing annotation: {e}")
        return []  # If parsing fails, return an empty list

# Apply parsing function to 'annotations' column
df['annotations_parsed'] = df['annotations'].apply(parse_annotation_str)

# Print out the parsed annotations to inspect the result
print(df['annotations_parsed'].head(10))  # Check if parsing works

# Extract relevant annotations (Life Stage, Sex, Alive or Dead)
def extract_individual_annotations(annotations):
    # Maps for controlled attribute and value IDs
    attribute_map = {
        1: 'Life Stage',
        9: 'Sex',
        17: 'Alive or Dead'
    }

    value_map = {
        2: 'Adult', 7: 'Egg', 8: 'Juvenile',
        10: 'Female', 11: 'Male',
        18: 'Alive', 19: 'Dead', 20: 'Cannot Be Determined'
    }

    # Initialize a dictionary to store extracted attributes
    extracted_data = {
        'life_stage': None,
        'sex': None,
        'alive_or_dead': None
    }

    # Check if annotations is a valid list
    if isinstance(annotations, list):
        # Loop through annotations to fill the extracted data
        for a in annotations:
            attr_id = a.get('controlled_attribute_id')
            val_id = a.get('controlled_value_id')

            # Assign values to corresponding columns based on the attribute ID
            if attr_id == 1 and val_id in value_map:
                extracted_data['life_stage'] = value_map[val_id]
            elif attr_id == 9 and val_id in value_map:
                extracted_data['sex'] = value_map[val_id]
            elif attr_id == 17 and val_id in value_map:
                extracted_data['alive_or_dead'] = value_map[val_id]

    return extracted_data

# Apply the extraction function to the 'annotations_parsed' column
annotation_columns = df['annotations_parsed'].apply(extract_individual_annotations)

# Expand the extracted data into separate columns
annotation_columns_df = annotation_columns.apply(pd.Series)

# Join the new columns with the original DataFrame
df = df.join(annotation_columns_df)

# Check if the new columns are populated
print(df[['id', 'life_stage', 'sex', 'alive_or_dead']].head())

# Optional: save to new CSV
df.to_csv("INat_Observations_NJ.csv_lifestage.csv", index=False)
print("Saved to INat_Observations_NJ.csv_lifestage.csv")
```

Analysis:

Intro: 
This analysis explores at observation information gathered from INaturalist database. This data is used to provide insights on behaviors of species that can be potentially impacted by the construction planned for the summer of 2025. The data provided includes observations from 2020 to 2024 and includes three species. The species included are the Diamondback terrapin, Mud fiddler crab (Minuca pugnax), and the Ipswich Sparrow (Passerculus sandwichensis princeps). The Terrapin was the main focus of the study, and the Mud fiddler was chosen because it is a food source for the Terrapin. The Ipswich Sparrow was chosen because it has little relation to the Terrapin in terms of species. This will provide insight into the effects on different animals. 

Data:
The Terrapin data shows that over the years the sightings have been inconsistent in growth. They are normally seen in June which is most likely when the females come up to nest. Most of the observations were of adults although there were some Juvenile ones seen as well. The data for the Mud fiddler crab shows that they are not spotted as much in the winter but besides that they have sighting in all other seasons. They have seen growth over the years and have only been observed as adults. The increase in sighting can be attributed to climate change. Lastly, the Ipswich Sparrow does not have many sightings and seems primarily in the winter seasons. 

results and conclusion: 
After reviewing the data, there are some concerns that the construction project brings forth. Addressing the Terrapins first, June is the month where most of the Terrapins are seen on land. Adding a bike path would mean another obstacle for the turtles to cross. On top of that it would be the middle of June which would mean an increase in human traffic due to the summer season. This would result in more human interaction with the turtles which could lead to more turtles being hit or other adverse outcomes that would otherwise be avoided if there was no bike path. Secondly the construction would take place in the summer, which would completely interrupt the nesting season for the species. The issues with the Mud fiddler mirror that of the Terrapin. They are summer/fall time creatures so the construction would cause a disruption in their natural cycle. Their population is also growing, and since they are a food source for the Terrapin as well as other animals in the area any disruptions to them will have an impact on other species as well. The Ipswich Sparrow are a wintertime animal so the construction itself will not impact them as much. However, they already have low number in the area so any disruptions to the eco system may cause them to drop off completely. Due to all of the reasons stated I would not recommend moving forward with the project if the desired outcome is to have little impact on the surrounding ecosystems.

The data may have possible errors due to input error. This could be an error in location input if someone did not put that their observation happened in NJ they would not be included in the numbers. Mispelling can also cause errors.
